## Deliverable C: Running 5 operations with numerical variables using the package `numpy`

I am intending to buy a 3 room AMK flat in Jan 2020 which costs $268000 and I want to get a house loan. 
So, I am trying to compare which type of SIBOR should I get, 1 month or 3 months.

The complete home loan rate is then calculated by adding the bankâ€™s charges (the fixed spread) to the SIBOR rate. 
Hence, assuming that the fixed spread is 0.70%. 
The below is the total interest that determines the monthly payments.

The 1 Month SIBOR rate is 1.749%.
The loaning bank has a fixed spread of 0.70%.
The total interest for the month: 1.749% (1 Month SIBOR) + 0.70% (fixed spread) = 2.449%

```python
price = 268000
interest_rate1 = 0.02449
years = 30
payments_year = 12

pmt1 = -1*np.pmt(0.02449/12, 30*12, 268000)
print(pmt1)
```

```python
mortgage1 = np.arange(0,3,dtype=np)
mortgage1[0] = price 
mortgage1[1] = interest_rate1
mortgage1[2] = np.round(pmt1, decimals = 2)

print(mortgage1)
```

The 3 Month SIBOR rate is 1.774%.
The loaning bank has a fixed spread of 0.70%.
The total interest for the month: 1.774% (3 Month SIBOR) + 0.70% (fixed spread) = 2.474%

```python
price = 268000
interest_rate2 = 0.02474
years = 30
payments_year = 12

pmt2 = -1*np.pmt(0.02474/12, 30*12, 268000)
print(pmt2)
```

```python
mortgage2 = np.arange(0,3,dtype=np)
mortgage2[0] = price 
mortgage2[1] = interest_rate2
mortgage2[2] = np.round(pmt2, decimals = 2)

print(mortgage2)
```

```python
difference = np.subtract((pmt2),(pmt1))

np.round(difference, decimals = 2)
```

Difference in Monthly Payment: $3.47

In this example, because the house loan is only $268000, hence the difference in monthly payment is rather small. If the house loan amoount is larger, there will be a larger difference.

Hence, as shown, the 3 month SIBOR has a higher monthly payment than the 1 month SIBOR but one has to take note of the current direction of interest rates. When SIBOR rates are falling, a 1 month rate will be cheaper than a 3 month rate. However, if interest rates are rising, the opposite is true.

1 month SIBOR rates adjust very quickly, whereas a 3 month rate takes time to adjust. So if SIBOR is falling month-on-month, the 1 month SIBOR would match the plunge, causing one's home loan to cost less.

But if SIBOR is rising, the 1 month rate would match it and rise just as quickly. Conversely, a 3 month SIBOR would maintain the lower rates from previous months.

## Deliverable G: Estimating 3 regression models using the package `statsmodels`

**Regression 1: Identifying the effect of floor area on the average price of resale HDB flats**

```python
d2reg1a = smf.ols("resale_price ~ floor_area_sqm", data = d2).fit()
print(d2reg1a.params)
```

We can interpret the following from the results:

_**For any additional square meter, the average price of resale HDB flats is increasing by $3988**_

```python
d2reg1b = smf.ols("np.log(resale_price) ~ np.log(floor_area_sqm)", data = d2).fit()
print(d2reg1b.params)
```

We can interpret the following from the results:

_**For any additional 1% larger floor area, the average price of resale HDB flats is increasing by 0.89%**_

```python
alpha = 49963.331048
beta = 3988.092224

d2['FitPrice'] = alpha + beta*d2['floor_area_sqm']

fig, ax = plt.subplots(1,1)
plt.scatter(d2['floor_area_sqm'],d2['resale_price'],s=4)
plt.plot(d2['floor_area_sqm'],d2['FitPrice'],color=(0,0,0))
plt.title('Price increase for 1 additional $m^2$')
plt.xlabel('Floor area ($m^2$)')
plt.ylabel('Transaction price (S$ 1 millions)')
plt.legend(['Linear fit','Actual transactions'])
plt.grid(color=(.9,.9,.9))
plt.show()
```

```python
d2reg1b.conf_int()
```

We can interpret the following from the results:
#### 95% Confidence Interval
_**With a 95% probability, the effect of the floor area on the average price of resale HDBs is between 0.88% and 0.90%**_

Moreover, we can also identify that these areas that corresponds to the highest HDB resale/rental prices are located near the Central/in the Central area itself. This means that people are often more inclined to purchase a property that is nearer to their jobs.
